```python
import pandas as pd

df = pd.read_csv("../data/raw/eth_food_price.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
df.columns = [
    'country', 'region', 'sub_region', 'market',
    'commodity', 'price_type', 'date',
    'frequency', 'price', 'unit', 'currency',
    'data_source', 'trend', 'pewi', 'alps_phase',
    'data_type', 'upper_ci', 'lower_ci', 'forecast_method'
]
```

```python
df = df[['country', 'region', 'market', 'commodity', 'price_type', 'date', 'price']]
```

```python
df.head()
```

```python
df = df[df["country"] == 'Ethiopia']
df = df[df['price_type'] == "Retail"]
```

```python
df.head()
df.info()
```

```python
df['date'] = pd.to_datetime(df['date'])
df['price'] = pd.to_numeric(df['price'], errors='coerce')
```

```python
df.isnull().sum()
```

```python
df = df.drop_duplicates()
```

```python
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
```

```python
df.describe()
df['commodity'].value_counts()
```

```python
df.to_csv("../data/processed/cleaned.csv", index=False)
```
